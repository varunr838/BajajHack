```python
import fitz  # PyMuPDF

def extract_text_from_pdf(file_path):
    doc = fitz.open(file_path)
    return "\n".join([page.get_text() for page in doc])
```

```python
def chunk_text(text, chunk_size=200, overlap=100):
    words = text.split()
    chunks = []
    for i in range(0, len(words), chunk_size - overlap):
        chunk = " ".join(words[i:i + chunk_size])
        chunks.append(chunk)
    return chunks
```

```python
text = extract_text_from_pdf('PDF1.pdf')
chunks = chunk_text(text)
for i in range(len(chunks)):
    print(f"Chunk {i + 1}: {chunks[i]}")
```

## **smart chunking**

```python
import fitz  # PyMuPDF
import re
from collections import defaultdict

def get_font_info(doc):
    """
    Analyzes font sizes across the document to establish a baseline for heading detection.
    Returns a dictionary of font sizes and their counts.
    """
    font_sizes = defaultdict(int)
    for page in doc:
        blocks = page.get_text("dict")["blocks"]
        for block in blocks:
            if block['type'] == 0: # text block
                for line in block['lines']:
                    for span in line['spans']:
                        font_sizes[round(span['size'], 1)] += 1 # Round to avoid float precision issues

    # Sort by frequency to find common body text size, or just get unique sizes
    unique_sizes = sorted(font_sizes.keys(), reverse=True)
    return unique_sizes

def is_potential_heading(span, unique_font_sizes):
    """
    Generalized heuristic to identify potential headings based on font size and boldness.
    Returns a 'level' (e.g., 1 for main, 2 for sub) or None.
    """
    font_size = round(span['size'], 1)
    is_bold = bool(span['flags'] & 2) # Flag 2 usually indicates bold

    # Simple heuristic: Larger fonts are higher-level headings
    # You might need to adjust these thresholds based on typical document structures
    if not unique_font_sizes:
        return None # Cannot determine without font size info

    if font_size >= unique_font_sizes[0] and is_bold: # Largest font, bold
        return 1
    elif len(unique_font_sizes) > 1 and font_size >= unique_font_sizes[1] and is_bold: # Second largest, bold
        return 2
    elif is_bold and font_size > unique_font_sizes[-1]: # Any bold text larger than smallest
        return 3

    return None

def smart_chunk_pdf_generalized(pdf_path, max_chunk_words=250, overlap_words=50):
    doc = fitz.open(pdf_path)
    chunks = []

    unique_font_sizes = get_font_info(doc)

    # Context stack: stores (level, header_text) tuples
    # Level 0 for document title, Level 1 for Section, Level 2 for Part/Major Point
    context_stack = []

    for page_num, page in enumerate(doc):
        blocks = page.get_text("dict")["blocks"]
        blocks.sort(key=lambda b: b['bbox'][1]) # Sort by Y-coordinate for reading order

        for block in blocks:
            if block['type'] == 1: # Image block, skip for text content
                continue

            block_lines = []
            for line in block['lines']:
                line_text = ""
                # Check font info for the first span of the line to classify it
                if line['spans']:
                    first_span = line['spans'][0]
                    line_text = first_span['text']

                    header_level = is_potential_heading(first_span, unique_font_sizes)

                    if header_level is not None:
                        # Update context stack
                        # Pop any lower or equal level headers
                        while context_stack and context_stack[-1][0] >= header_level:
                            context_stack.pop()
                        context_stack.append((header_level, line_text.strip()))
                        continue # This line is a header, not content

                # If it's not identified as a header, add to block_lines
                block_lines.append(line_text)

            block_text = "\n".join(block_lines).strip()

            if not block_text:
                continue

            # Prepend current hierarchical context
            current_context_prefix = " > ".join([h_text for level, h_text in context_stack])
            if current_context_prefix:
                chunk_content = f"{current_context_prefix}\n{block_text}"
            else:
                chunk_content = block_text

            # Simple word-based splitting for chunking.
            # For production, consider a more robust text splitter that handles sentences/paragraphs.
            words = chunk_content.split()

            # Generate chunks with overlap
            for i in range(0, len(words), max_chunk_words - overlap_words):
                chunk_words = words[i : i + max_chunk_words]
                chunk_text = " ".join(chunk_words)

                chunks.append({
                    "text": chunk_text,
                    "metadata": {
                        "page_number": page_num + 1,
                        "context_path": current_context_prefix, # Store the full path for metadata
                        "section": context_stack[0][1] if context_stack else None, # Main section
                        "subsection": context_stack[1][1] if len(context_stack) > 1 else None, # Sub-section
                        # You can add more specific metadata based on your `context_stack` levels
                    }
                })

    doc.close()
    return chunks

# --- Example Usage ---
# Create a dummy PDF for testing (using your previous content)

dummy_pdf_path = "PDF1.pdf"

# Generate chunks
chunks = smart_chunk_pdf_generalized(dummy_pdf_path, max_chunk_words=150, overlap_words=30)
```

```python
import chromadb
from sentence_transformers import SentenceTransformer

# 1. Create Chroma DB
client = chromadb.Client()
try:
    client.delete_collection(name="my_docs")
    print("Deleted existing collection 'my_docs'")
except:
    print("Collection 'my_docs' did not exist")

collection = client.create_collection(name="my_docs")


# 2. Embed some chunks
model = SentenceTransformer("all-MiniLM-L6-v2")
text = extract_text_from_pdf('PDF1.pdf')
chunks_with_metadata = chunks  # Use the chunks generated by smart_chunk_pdf_generalized
texts = [chunk["text"] for chunk in chunks_with_metadata] # Extract only the text
embeddings = model.encode(texts).tolist()

# Filter out None values from metadata
metadatas = []
for chunk in chunks_with_metadata:
    cleaned_metadata = {k: v for k, v in chunk["metadata"].items() if v is not None}
    metadatas.append(cleaned_metadata)


# 3. Add to Chroma
ids = [str(i) for i in range(1, len(texts) + 1)]
collection.add(documents=texts, embeddings=embeddings, ids=ids, metadatas=metadatas)
```

```python
# 4. Embed a user query and search
query = "I'm planning to get an eye examination for my refractive error. Is this covered?"
query_embedding = model.encode([query]).tolist()[0]

results = collection.query(query_embeddings=[query_embedding], n_results=2)
print(results['documents']," ",results['metadatas'])
```
